```python
import pandas as pd
import quandl
import os
from dotenv import load_dotenv
from quandl.errors import quandl_error
```

```python
load_dotenv()
```

```python
quandl.ApiConfig.api_key =  os.getenv("QUANDL_API_KEY")
```

```python
# get symbols of current secrities in the S&P 500
table=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
SP500_df = table[0]
SP500_df.head()
```

```python
# for each symbol retrieve EOD market information from QUANDL
EOD_DF = pd.DataFrame()
for symbol in SP500_df['Symbol']:
    symbol = symbol.replace('.','_') #need to replace . wiith _ to match symbol on quandl
    try: 
        mydata = quandl.get(f"EOD/{symbol}")
        print(f'Pulling data for {symbol}')
        EOD_DF[f'{symbol}_closing'] = mydata['Adj_Close']
    except quandl_error.ForbiddenError:
        print(f'need premission for {symbol}')
EOD_DF        
```
